<h1>Telecom Chrun Prediction</h1>


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("Telco-Customer-Churn.csv")
print(data)
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='Churn', y='tenure', data=data, palette='Set2')
plt.title("Customer Tenure by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Tenure (months)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap")
plt.show()
```

<h2>Cleaning Data</h2>


```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

if 'customerID' in data.columns:
    data.drop('customerID', axis=1, inplace=True)

data['Churn'] = data['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].median())

label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

scaler = StandardScaler()
scaled_features = scaler.fit_transform(data.drop('Churn', axis=1))
```

```python

X_train,X_test,y_train,y_test=train_test_split(
    scaled_features, data['Churn'],test_size=0.2,random_state=42)

plt.figure(figsize=(6,4))
sns.countplot(x=y_train, palette='Set1')
plt.title("Churn Distribution - Before Balancing")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

print("Training data",X_train.shape)
print("Testing data",X_test.shape)
print("\nAfter balancing:")
print(pd.Series(y_train).value_counts())



# AFTER balancing
plt.figure(figsize=(6,4))
sns.countplot(x=y_train, palette='Set1')
plt.title("Churn Distribution - After Balancing with SMOTE")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(n_estimators=200, random_state=42)
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
rf_acc= accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
feature_importances = pd.Series(model2.feature_importances_, index=data.drop('Churn', axis=1).columns)
top_features = feature_importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
top_features.plot(kind='barh')
plt.title("Top 10 Feature Importances (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.gca().invert_yaxis()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

y_pred_proba = model2.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Random Forest')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
plt.figure(figsize=(6,5))
plt.plot(recall, precision, color='purple')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```
